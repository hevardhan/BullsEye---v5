```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv("../EURUSD(2023).csv")
```

```python
data
```

```python
def calculate_macd(data, short_period=12, long_period=26, signal_period=9):
    """Calculate the MACD and signal line."""
    data['ema_short'] = data['close'].ewm(span=short_period, adjust=False).mean()
    data['ema_long'] = data['close'].ewm(span=long_period, adjust=False).mean()

    data['MACD'] = data['ema_short'] - data['ema_long']
    data['signal_line'] = data['MACD'].ewm(span=signal_period, adjust=False).mean()

    data['MACD_histogram'] = data['MACD'] - data['signal_line']

    return data
```

```python
df = calculate_macd(data, short_period=12, long_period=26, signal_period=9)

df
```

```python
df.drop(['Unnamed: 0','spread','tick_volume','real_volume'],axis=1,inplace=True)
```

```python
df.head(10)
```

```python
import pandas as pd

# Assuming df is your DataFrame
df['signal'] = 0  # Initialize all with 0 (Sell)

# Condition for Buy signal: MACD_histogram crosses from below to above 0
df['signal'] = df['MACD_histogram'].apply(lambda x: 1 if x > 0 else 0)

# Or if you want to consider only crossovers:
df['signal'] = (df['MACD_histogram'] > 0) & (df['MACD_histogram'].shift(1) <= 0)
df['signal'] = df['signal'].astype(int)  # Convert boolean to 1/0

# Display the DataFrame with the new signal column
print(df)
```
